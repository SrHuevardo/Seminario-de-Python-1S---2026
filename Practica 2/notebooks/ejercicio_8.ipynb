```python
import sys
sys.path.append('../src')

import caesar_cipher as cc
```

```python
mensaje_ingresado = input("Ingrese el mensaje a cifrar: ")

valor_desplazamiento = int(input("Ingrese el valor de desplazamiento (número): "))

mensaje_cifrado = cc.apply_cipher(mensaje_ingresado, valor_desplazamiento)
print(f"Mensaje cifrado:  {mensaje_cifrado}")

mensaje_descifrado = cc.apply_cipher(mensaje_cifrado, -valor_desplazamiento)
print(f"Mensaje devuelto: {mensaje_descifrado}")
print("-" * 40)
```
